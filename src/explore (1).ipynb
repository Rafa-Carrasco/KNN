```python
# Your code here

import requests
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split

# descargar data

url = "https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv"
nombre_archivo = "tmdb_5000_credits.csv"

respuesta = requests.get(url)

with open(nombre_archivo, 'wb') as archivo:
    archivo.write(respuesta.content)

url = "https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv"
nombre_archivo = "tmdb_5000_movies.csv"

respuesta = requests.get(url)

with open(nombre_archivo, 'wb') as archivo:
    archivo.write(respuesta.content)

```

```python
# convertir csv en dataframe

cre_df = pd.read_csv("../data/raw/tmdb_5000_credits.csv")
cre_df.shape
cre_df.info() 

mov_df = pd.read_csv("../data/raw/tmdb_5000_movies.csv")
mov_df.shape
mov_df.info() 
```

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('workraf.db')

cre_df.to_sql('credits', conn, if_exists='replace', index=False)
mov_df.to_sql('movies', conn, if_exists='replace', index=False)

cur = conn.cursor()

# cur.execute('''
#         DROP TABLE merged_table
# ''')
            
cur.execute('''
   CREATE TABLE merged_table AS
    SELECT *
    FROM credits
    INNER JOIN movies ON credits.title = movies.title
''')

total_data = pd.read_sql('SELECT * FROM merged_table', conn)
conn.close()

total_data.head()

```

```python

total_data_df = total_data.drop(['budget', 'homepage', 'id', 'original_language', 'original_title',
       'popularity', 'production_companies', 'production_countries',
       'release_date', 'revenue', 'runtime', 'spoken_languages', 'status',
       'tagline', 'title:1', 'vote_average', 'vote_count'], axis=1)
total_data_df.head()
```

```python
import json
import pandas as pd


def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
    
total_data_df['genres'] = total_data_df['genres'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
total_data_df['keywords'] = total_data_df['keywords'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
total_data_df["cast"] = total_data_df["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)
total_data_df["crew"] = total_data_df["crew"].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))
total_data_df["overview"] = total_data_df["overview"].apply(lambda x: [x])

total_data_df.head()

```

```python
total_data_df["overview"] = total_data_df["overview"].apply(lambda x: [str(x)])
total_data_df["genres"] = total_data_df["genres"].apply(lambda x: [str(genre) for genre in x])
total_data_df["keywords"] = total_data_df["keywords"].apply(lambda x: [str(keyword) for keyword in x])
total_data_df["cast"] = total_data_df["cast"].apply(lambda x: [str(actor) for actor in x])
total_data_df["crew"] = total_data_df["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

total_data_df["tags"] = total_data_df["overview"] + total_data_df["genres"] + total_data_df["keywords"] + total_data_df["cast"] + total_data_df["crew"]
total_data_df["tags"] = total_data_df["tags"].apply(lambda x: ",".join(x).replace(",", " "))

total_data_df.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)
total_data_df.iloc[0].tags
```

```python
total_data_df.to_csv("../data/processed/clean_data.csv", index = False)

conn = sqlite3.connect("../data/workraf.db")
mov_df.to_sql("clean_movies_data", conn, if_exists = "replace", index = False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_vector = tfidf_vectorizer.fit_transform(total_data_df['tags'])
print("\nTF-IDF Vectorizer:")
print(tfidf_vector.toarray())
print(tfidf_vectorizer.get_feature_names_out())

```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors( n_neighbors=6, algorithm="brute", metric="cosine")
model.fit(tfidf_vector)

```

```python
distancias, indices = model.kneighbors(tfidf_vector)
print("Distancias:", distancias)
print("Índices de los vecinos más cercanos:", indices)

for indice in indices[0]:
    print(tfidf_vector[indice])
```

```python
def get_movie_recommendations(movie_title):
    movie_index = total_data_df[total_data_df["title"] == movie_title].index[0]
    distances, indices = model.kneighbors(tfidf_vector[movie_index])
    similar_movies = [(total_data_df["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "Avatar"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
```
